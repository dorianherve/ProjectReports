```python
from numpy import *
from random import *
import scipy.stats
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import matplotlib.pyplot as plt3
import matplotlib.pyplot as plt4
```

## Code pour les simulations des trajectoires et des convergences p.s.

### Dans le cas sans renforcement :

```python
def simulation_trajectoire_sans_renforcement(n):

    #déclaration des variables
    tab_abs=[]
    tab_ord=[]
    A = []
    B = []
    X = []
    Y=[]
    Z=[]
    limY = []
    limZ = []

    #initialisation
    A.append(1)
    B.append(1)
    X.append(0)
    Y.append(0)
    Z.append(0)


    for i in range (0,n) :

        #loi uniforme [0,1]
        U=random()
        tab_abs.append(i)

        #chemin alpha
        if(U<=A[i]/(A[i]+B[i])):
           X.append('a')
           tab_ord.append(4)
           A.append(A[i])
           B.append(B[i])
           #nb de passages sur alpha augmente de 1
           Y.append(Y[i]+1)
           Z.append(Z[i])
           
        #chemin beta
        else :
            X.append('b')
            tab_ord.append(3)
            A.append(A[i])
            B.append(B[i])
            Y.append(Y[i])
            #nb de passages sur beta augmente de 1
            Z.append(Z[i]+1)
            
        if(i==0):
            limY.append(0)
            limZ.append(0)
        else :
            limY.append(Y[i]/i)
            limZ.append(Z[i]/i)


    print('Proportion passages sur alpha :',Y[n]/n)
    print('Proportion passages sur beta :',Z[n]/n)

    #évolution du choix de chemin
    plt.scatter(tab_abs,tab_ord)
    plt.title("Trajectoire Xn dans le cas sans renforcement")
    plt.xlabel('Individu')
    plt.ylabel('Chemin')
    plt.show()

    
    #limite ps de Yn/n
    plt2.scatter(tab_abs,limZ,s=5,c='darkorange')
    plt2.title("Convergence p.s. de Yn/n dans le cas sans renforcement")
    plt2.xlabel("Individu")
    plt2.ylabel("Yn/n")
    plt2.plot([0.0,n], [1/2,1/2], c='black')
    plt2.show()
    
    #limite ps de Zn/n
    plt2.scatter(tab_abs,limZ,s=5,c='darkorange')
    plt2.title("Convergence p.s. de Zn/n dans le cas sans renforcement")
    plt2.xlabel("Individu")
    plt2.ylabel("Zn/n")
    plt2.plot([0.0,n], [1/2,1/2], c='black')
    plt2.show()
    
    


n=input('Entrez la valeur de n=');
n=int(n)

simulation_trajectoire_sans_renforcement(n)
           

```

## Code pour les simulations des convergences en loi

### Dans le cas sans renforcement :

```python
def simulation_trajectoire_sans_renforcement(n,mode):

    #déclaration des variables
    tab_abs=[]
    tab_ord=[]
    A = []
    B = []
    X = []
    Y=[]
    Z=[]
    limY = []
    limZ = []

    #initialisation
    A.append(1)
    B.append(1)
    X.append(0)
    Y.append(0)
    Z.append(0)

    limY.append(0)
    limZ.append(0)

    for i in range (0,n) :

        #loi uniforme [0,1]
        U=random()
        tab_abs.append(i)

        #chemin alpha
        if(U<=A[i]/(A[i]+B[i])):
           X.append('a')
           tab_ord.append(4)
           A.append(A[i])
           B.append(B[i])
           #nb de passages sur alpha augmente de 1
           Y.append(Y[i]+1)
           Z.append(Z[i])
           
        #chemin beta
        else :
            X.append('b')
            tab_ord.append(3)
            A.append(A[i])
            B.append(B[i])
            Y.append(Y[i])
            #nb de passages sur beta augmente de 1
            Z.append(Z[i]+1)
            
        if(i==0):
            limY.append(0)
            limZ.append(0)
        else :
            limY.append(Y[i]/i)
            limZ.append(Z[i]/i)
  

    #1 = Yn
    if(mode==1):
        return 2*sqrt(n)*(Y[n]/n - 1/2)
    else:
        return 2*sqrt(n)*(Z[n]/n - 1/2)



def simulation_cv_loi_sans_renforcement(N,n,mode):
    tab_loi = []
    for i in range (0,N):
        l = simulation_trajectoire_sans_renforcement(n,mode)
        tab_loi.append(l)

    mu = 0
    sigma = 1

    n, bins, _ = plt.hist(tab_loi, bins = 40, normed=1, color='cadetblue', edgecolor='darkslategrey')
    y1 = scipy.stats.norm.pdf(bins, mu, sigma)

    plt.ylabel('f(x)')
    plt.xlabel('x')
    if(mode==1):
        plt.title('Convergence en loi de Yn dans le cas sans-renforcement')
    else:
        plt.title('Convergence en loi de Zn dans le cas sans-renforcement')

    plt.plot(bins,y1,'red')
    plt.show()

    
n=input("Entrez la valeur de n (nombre d'individus) = ");
n=int(n)

N=input("Entrez la valeur de N (nombre d'expériences) = ");
N=int(N)

mode=input('Choisissez le mode (1 : convergence de Yn et 2 : convergence de Zn) =');
mode=int(mode)
    
#1 = convergence en loi de Yn , 2 = convergence en loi de Zn
simulation_cv_loi_sans_renforcement(N,n,mode)

```

### Dans le cas du renforcement linéaire:

Pour calculer : 
On va calculer U avec un n=5000 et pour $\frac{Y_{n}}{n}$ on prend un n plus petit, ici n=4100.

```python
def simulation_trajectoire_renforcement_lineaire(n,mode,petit_n):

    #déclaration des variables
    tab_abs=[]
    tab_ord=[]
    A = []
    B = []
    X = []
    Y=[]
    Z=[]
    limY = []
    limZ = []

    #initialisation
    A.append(1)
    B.append(1)
    X.append(0)
    Y.append(0)
    Z.append(0)
    tab_abs.append(0)
    tab_ord.append(0)
    yn_sur_n = 0
    zn_sur_n = 0

    for i in range (0,n) :

        #loi uniforme [0,1]
        U=random()
        
        tab_abs.append(i)

        #chemin alpha
        if(U<=A[i]/(A[i]+B[i])):
           X.append('a')
           tab_ord.append(4)
           A.append(A[i]+1)
           B.append(B[i])
           #nb de passages sur alpha augmente de 1
           Y.append(Y[i]+1)
           Z.append(Z[i])
          
        #chemin beta
        else :
            X.append('b')
            tab_ord.append(3)
            A.append(A[i])
            B.append(B[i]+1)
            Y.append(Y[i])
            #nb de passages sur beta augmente de 1
            Z.append(Z[i]+1)
          
        if(i==0):
            limY.append(0)
            limZ.append(0)
        else :
            limY.append(Y[i]/i)
            limZ.append(Z[i]/i)   

        if (i==petit_n):
            yn_sur_n = Y[i]/i
            zn_sur_n = Z[i]/i

    
    if (mode==1):
        return 6*sqrt(n)*(yn_sur_n - Y[n]/n ) 
    else:
        return 6*sqrt(n)*(zn_sur_n - Z[n]/n ) 




def simulation_cv_loi_renforcement_lineaire(N,n,mode,petit_n):
    tab_loi = []
    for i in range (0,N):
        l = simulation_trajectoire_renforcement_lineaire(n,mode,petit_n)
        tab_loi.append(l)

    mu = 0
    sigma = 1

    n, bins, _ = plt.hist(tab_loi, bins = 40, normed=1, color='cadetblue', edgecolor='darkslategrey')
    y1 = scipy.stats.norm.pdf(bins, mu, sigma)

    plt.ylabel('f(x)')
    plt.xlabel('x')
    if(mode==1):
        plt.title('Convergence en loi de Yn dans le cas renforcement linéaire')
    else:
        plt.title('Convergence en loi de Zn dans le cas renforcement linéaire')

    plt.plot(bins,y1,'red')
    plt.show()

print("Recommandé : n=5000, N=10000, mode= 1 ou 2, petit n= 4100 \n")    
n=input("Entrez la valeur de n (nombre d'individus) = ");
n=int(n)

N=input("Entrez la valeur de N (nombre d'expériences) = ");
N=int(N)

mode=input('Choisissez le mode (1 : convergence de Yn et 2 : convergence de Zn) = ');
mode=int(mode)

petit_n=input("Entrez la valeur du petit n choisi pour Yn/n (plus petit que le nombre d'individus) = ");
petit_n=int(petit_n)

#choix du n plus petit pour la convergence en loi
petit_n=petit_n

#1 = convergence en loi de Yn , 2 = convergence en loi de Zn
simulation_cv_loi_renforcement_lineaire(N,n,mode,petit_n)


```
